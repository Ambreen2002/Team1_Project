```python
def find_beta(df):

    index_returns = df.iloc[:,0]
    df_beta = {}
    
    for ticker in df:
        covariance = df[ticker].cov(index_returns)
        variance = df[ticker].var()
        beta = covariance/variance
        df_beta[ticker] = beta
    
    return df_beta
```
